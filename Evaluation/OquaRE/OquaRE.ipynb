```python
def RadialGraph(subchar:[], score1:[], scores:[], title:str, legend:str, pos=[0.2,0.2]):
    score1 = [*score1, score1[0]]
    #score2 = [*score2, score2[0]]
    # Adding first element for radial graph
    new_scores = []
    for score in scores:
        new_scores.append([*score, score[0]])
    subchar = [*subchar, subchar[0]]
    # Initialise the spider plot by setting figure size and polar projection
    plt.figure(figsize=(10, 5))
    plt.subplot(polar=True)    
    theta = np.linspace(start = 0, stop = 2 * np.pi, num = len(score1))
  
    # Arrange the grid into number of sales equal parts in degrees
    #lines, labels = plt.thetagrids(range(0, 360, int(360/len(subchar))), (subchar))
    lines, labels = plt.thetagrids(np.degrees(theta), labels = subchar)

    # Plot onto1
    plt.plot(theta, score1)
    plt.fill(theta, score1, 'b', alpha=0.1)
    
    # Plot onto2
    for score in new_scores:
        plt.plot(theta, score, linestyle = '--')
    #plt.fill(theta, score2, 'b', alpha=0.1)
    
    # Add legend and title for the plot
    leg = plt.legend(labels=(legend), loc=1 , bbox_to_anchor=(pos))
    plt.title(title)
   
    # Display
    plt.show()
```
