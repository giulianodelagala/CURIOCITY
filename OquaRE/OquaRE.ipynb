```python
def RadialGraph(subchar:[], score1:[], score2:[], title:str, legend:str, pos=[0.2,0.2]):
    # Initialise the spider plot by setting figure size and polar projection
    plt.figure(figsize=(10, 5))
    plt.subplot(polar=True)    
    theta = np.linspace(0, 2 * np.pi, len(score1))
  
    # Arrange the grid into number of sales equal parts in degrees
    lines, labels = plt.thetagrids(range(0, 360, int(360/len(subchar))), (subchar))
    
    # Plot onto1
    plt.plot(theta, score1)
    plt.fill(theta, score1, 'b', alpha=0.1)
    
    # Plot onto2
    plt.plot(theta, score2)
    #plt.fill(theta, score2, 'b', alpha=0.1)
    
    # Add legend and title for the plot
    leg = plt.legend(labels=(legend), loc=1 , bbox_to_anchor=(pos))
    plt.title(title)
   
    # Display
    plt.show()
```

```python
metric_char = ['LCOMOnto', 'WMCOnto', 'DITOnto', 'NACOnto', 'NOCOnto', 'CBOOnto', 'RFCOnto', 'NOMOnto', 'RROnto', 'INROnto', 'TMOnto']
cit_score = [E.LCOMOnto[1], E.WMCOnto2[1], E.DITOnto[1], E.NACOnto[1], E.NOCOnto[1], E.CBOOnto[1], E.RFCOnto[1], E.NOMOnto[1], E.RROnto[1], E.INROnto[1], E.TMOnto2[1]] 
erl_score = [CRM.LCOMOnto[1], CRM.WMCOnto2[1], CRM.DITOnto[1], CRM.NACOnto[1], CRM.NOCOnto[1], CRM.CBOOnto[1], CRM.RFCOnto[1], CRM.NOMOnto[1], CRM.RROnto[1], CRM.INROnto[1], CRM.TMOnto2[1] ] 
df = pd.DataFrame({'value1':cit_score , 'value2':erl_score })
 
my_range=range(1,len(df.index)+1)
 
# The horizontal plot is made using the hline function
plt.hlines(y=my_range, xmin=df['value1'], xmax=df['value2'], color='grey', alpha=0.4)
plt.scatter(df['value1'], my_range, color='red', alpha=1, label='CURIOCITY', s=80)
plt.scatter(df['value2'], my_range, color='blue', alpha=0.4 , label='ERLANGEN-CRM', s=80, marker='s')
plt.legend(loc=8, bbox_to_anchor=[0.5,-0.35])
plt.rcParams['grid.alpha'] = .1
plt.grid() 
# Add title and axis names
plt.yticks(my_range, metric_char)
plt.title("Structural Level Evaluation", loc='center')
plt.xlabel('OquaRE Score')
plt.ylabel('Metrics')

# Show the graph
plt.show()
```
